```r
df <- read.csv("/dsa/data/DSA-7030/Chicago-Crime-Sample-2012.csv",sep=",",strip.white = TRUE)
```

```r
head(df)
```

```r
suppressMessages(library(dplyr))

df <- rename(df, Primary_Type = Primary.Type)
df <- rename(df, Location_Description = Location.Description)
df <- rename(df, Community_Area = Community.Area)
df <- rename(df, X_Coordinate = X.Coordinate)
df <- rename(df, Y_Coordinate = Y.Coordinate)
df <- rename(df, Case_Number = Case.Number)
df <- rename(df, FBI_Code = FBI.Code)
df <- rename(df, Updated_On = Updated.On)
```

```r
case <- select(df, Case_Number, Arrest, FBI_Code,  Domestic, IUCR, Primary_Type, Description, Date, Block, Beat, District, Location_Description, Community_Area, Ward, X_Coordinate, Y_Coordinate, Latitude, Longitude)
record <- select(df, ID, Case_Number, Updated_On)
```
